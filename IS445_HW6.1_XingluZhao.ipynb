```python
import altair as alt
import pandas as pd

# Load and preprocess the dataset
url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/ufo-scrubbed-geocoded-time-standardized-00.csv"
data = pd.read_csv(url)
data.columns = ['date', 'location', 'state', 'country', 'shape', 'duration_seconds', 'duration_text', 'comments', 'date_posted', 'latitude', 'longitude']
data['year'] = pd.to_datetime(data['date'], errors='coerce').dt.year

# Filter data for sightings after 1950 and limit duration to a reasonable range
filtered_data = data[(data['year'] >= 1950) & (data['duration_seconds'] < 100000)].copy()
filtered_data['decade'] = (filtered_data['year'] // 10) * 10

# Disable max rows limit for Altair
alt.data_transformers.disable_max_rows()

# Create the Altair chart
chart = alt.Chart(filtered_data).mark_circle(opacity=0.3, size=50).encode(
    x=alt.X('decade:O', title='Decade of UFO Sighting', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('duration_seconds', scale=alt.Scale(type='log'), title='Duration of UFO Sighting (seconds)'),
    color=alt.Color('shape', title='UFO Shape', legend=alt.Legend(orient='right')),
    tooltip=['year', 'duration_seconds', 'location', 'shape']
).properties(
    title='UFO Sightings Duration by Decade (Log Scale, 1950+)'
).interactive()  # Enable zoom and pan

# Display the chart
chart.show()


#UFO Sightings Duration by Decade (Log Scale, 1950+)
#Visual Description: The first chart shows the distribution of UFO sighting durations (in seconds) for each decade starting in 1950. The UFO sightings for each decade are grouped using a bar chart and the durations are shown on a logarithmic scale in order to see the trend in the distribution of durations more clearly. The charts coded the different UFO shapes by their different shapes, allowing the user to observe the change in duration for each shape.
#Design choices: We used a logarithmic scale for the Y-axis in order to deal with very large data over a time span and to make the chart more readable. the X-axis represents the chronological age of the UFO sightings, and by grouping the time periods, it is possible to observe the trend of the duration of UFO sightings within each decade more intuitively. Colors are used to indicate the shape of the UFO and the legend is placed on the right side. Through these design choices, we aim to help users understand the differences in the duration of UFO sightings in each decade, and to explore the differences in the duration of different UFO shapes.
#Data Transformation: We filtered the data to select only sightings from 1950 and later, and limited the durations to less than 100,000 seconds. For better grouping, we grouped the years into units (decades) of 10 years to show trends over time.
#Interactivity: The chart allows the user to zoom and pan, which allows the user to view the data in more detail for specific time periods to enhance the explorability of the chart.


```

```python
import altair as alt
import pandas as pd

# Load and preprocess the dataset
url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/ufo-scrubbed-geocoded-time-standardized-00.csv"
data = pd.read_csv(url)
data.columns = ['date', 'location', 'state', 'country', 'shape', 'duration_seconds', 'duration_text', 'comments', 'date_posted', 'latitude', 'longitude']
data['year'] = pd.to_datetime(data['date'], errors='coerce').dt.year

# Filter data for sightings after 1950 and limit duration to a reasonable range, and exclude missing country values
filtered_data = data[(data['year'] >= 1950) & (data['duration_seconds'] < 100000) & (data['country'].notna())].copy()

# Disable max rows limit for Altair
alt.data_transformers.disable_max_rows()

# Create a selection for interaction using selection_point()
selection = alt.selection_point(fields=['country'], bind='legend', empty='all')

# Create the second Altair chart - UFO sightings by country
chart = alt.Chart(filtered_data).mark_bar().encode(
    x=alt.X('country:N', title='Country of UFO Sighting', sort='-y'),
    y=alt.Y('count()', title='Number of UFO Sightings'),
    color=alt.Color('country:N', title='Country'),
    tooltip=['country', 'count()']
).add_params(
    selection
).transform_filter(
    selection
).properties(
    title='UFO Sightings by Country (1950+)',
    width=600
).interactive()  # Enable zoom and pan

# Display the chart
chart.show()



#UFO Sightings by Country (1950+)
#Visual Description: The second chart shows the number of UFO sightings by country since 1950. We use a bar chart to show the number of UFO sightings by country, where the x-axis represents the country and the y-axis represents the number of sightings.
#DESIGN OPTION: In this graph, the X-axis is sorted by country and the Y-axis represents the number of UFO sightings. Each bar is color-coded for a different country to help users quickly distinguish UFO sightings in each country. The width of the bars is relatively large in order to effectively show the difference in numbers even in a limited space.
#Data Conversion: To ensure the validity of the data, we filtered the data from 1950 and later and removed data with missing country information. This ensures that the chart more accurately reflects UFO sightings in each country.
#Interactivity: The chart allows the user to zoom and pan so that they can better focus on the data for certain countries, thus enhancing the chart's explorability and comprehensibility.


```
